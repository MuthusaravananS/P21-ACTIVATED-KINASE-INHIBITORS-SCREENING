```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'263mordred3dinput.csv')
df.shape
df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
X = df.drop("pIC50", axis=1)
Y = df["pIC50"]
df.head(10)
df=df.apply(lambda x: pd.to_numeric(x, errors='coerce'))
df=df.replace(np.nan, 0)
X = df.drop("pIC50", axis=1)
y = df["pIC50"]
X = np.nan_to_num(X)
y = np.nan_to_num(y)

#Spliting the dataset into Training set and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=0)
```

```python
import matplotlib.pyplot as plt
from catboost import CatBoostRegressor
import sklearn.metrics as metrics
CATreg = CatBoostRegressor()
CATreg.fit(X_train,y_train)
CAT_cod=CATreg.score(X_test,y_test)

plt.scatter(y_train,CATreg.predict(X_train),c='green', label='Training')
plt.plot(y_train,y_train,c='black')
plt.scatter(y_test,CATreg.predict(X_test),c='red', label='Test')
plt.xlabel('Actual pIC50', fontweight='bold')
plt.ylabel('Predicted pIC50', fontweight='bold')
plt.legend(loc='upper right', numpoints=1, ncol=3, fontsize=8)

plt.show()


# Model prediction on train data
y_pred = CATreg.predict(X_train)
# Predicting Test data with the model
y_test_pred = CATreg.predict(X_test)


# Model Evaluation
print('Coefficient of determination(R^2) = {}'.format(CAT_cod)) 
print('MAE:',metrics.mean_absolute_error(y_test, y_test_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
print('MAPE',np.mean(np.abs(( y_test - y_test_pred)/y_test))*100)
```

```python
#XGbooost
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
xgb=XGBRegressor()
parameters = {
    'max_depth': range (2, 10, 1),
    'n_estimators': range(60, 220, 40),
    'learning_rate': [0.1, 0.01, 0.05]
}
gcv = GridSearchCV(
    estimator=xgb,
    param_grid=parameters,
    scoring = 'neg_root_mean_squared_error',
    n_jobs = -1,
    cv = 10,
    verbose=True
)

search=gcv.fit(X_train,y_train)
search.best_params_
```
